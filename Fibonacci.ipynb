**Method 1**

```python
def fibonacci(n):
  if n<=1:
    return n
  else:
    return(fibonacci(n-1)+fibonacci(n-2))
```

```python
  n=11
  for i in range(n):
    print(fibonacci(i),end=" ")
```

**Method 2**

```python
def fibonacci_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci_iterative(10)
```
